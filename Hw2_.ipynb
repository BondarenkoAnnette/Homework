```python
# необходимые импорты
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
```

### шаг 1

```python
# соберите все ссылки на страницы с новостями (цикл for + range)
# 1 страница: https://sysblok.ru
# 2 страница: https://sysblok.ru/page/2
# последняя: https://sysblok.ru/page/16

all_page = ['https://sysblok.ru', ]

# ваш код

url = "https://sysblok.ru/"


for i in range(2,17):
  all_page.append(url+"page"+str(i))
all_page
```

### шаг 2

```python
# с каждой страницы соберите ссылки на отдельные новости
# подсказка: родительский тег h2 (проверяйте по нему), нужный нам - a

all_links = []

# ваш код

page = rq.get(all_page[2]) #здесь я беру только новости с 3 страницы сайта
soup = BeautifulSoup(page.text, features="html.parser")

for link in soup.find_all('a'):
    if link.parent.name == "h2":
      all_links.append(link.get('href'))
all_links

```

### шаг 3

работаем с 1 новостью:

нас интересует название, автор, дата публикации, текст (дополнительно можно спарсить тематические категории)

```python
url0 = all_links[0]

page0 = rq.get(url0)
soup0 = BeautifulSoup(page0.text)
print(soup0.prettify())
```

```python
# для парсинга даты: извлеките текст из тега time
print(soup0.find('time').text)
```

```python
# для парсинга заголовка: извлеките текст из тега h1
print(soup0.find('h1').text)
```

```python
# для парсинга текста: соберите все тексты из тега p и соедините в строку
text_list = soup0.find_all('p')

text = []
for i in text_list:
  text.append(i.text)

final_text = ' '.join(text)
final_text = final_text.replace('\xa0', ' ')
final_text = final_text.replace('\n', '')
final_text
```

```python
# автора собрать сложнее:
# достаем из тега meta как на паре или из тега a с проверкой на содержимое атрибута rel

for i in soup0.find_all('meta'):
  if i.get('name') == 'author':
    author = i.get("content")

print(author)
```

```python
# тематические категории:
# <a rel="category tag"> нужные нам категории </a>
# можно попробовать так:
# if i.get('rel') == ['category', 'tag']

for i in soup0.find_all('a'):
  if i.get('rel') == ['category', 'tag']:
    category = i.get_text()
print(category)
```

### шаг 4

```python
# пишем функцию по сбору информации
def Getnews (url0):
  page0 = rq.get(url0)
  soup0 = BeautifulSoup(page0.text)
  date = soup0.find('time').text
  title = soup0.find('h1').text
  for i in soup0.find_all('meta'):
    if i.get('name') == 'author':
      author = i.get("content")
  for i in soup0.find_all('a'):
    if i.get('rel') == ['category', 'tag']:
      category = i.get_text()
  text_list = soup0.find_all('p')
  text = []
  for i in text_list:
    text.append(i.text)
  final_text = ' '.join(text)
  final_text = final_text.replace('\xa0', ' ')
  final_text = final_text.replace('\n', '')
  return date, title, author, category, final_text
```

```python
#проверка функции
Getnews (all_links[4])
```

```python
# собираем все новости!
# полезно использовать try / except

news = []

for link in all_links:
  try:
    new = Getnews(link)
    news.append(new)
  except:
    print(link)
```

### шаг 5

```python
# соберите всю собранную информацию в датафрейм
df = pd.DataFrame(news)
df.columns = ['дата публикации', 'заголовок статьи', 'автор', 'категория', 'текст']
df.head()
```
